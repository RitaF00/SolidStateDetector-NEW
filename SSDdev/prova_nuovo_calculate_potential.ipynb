```julia
using Statistics
```

```julia
println("Calcolo delle differenze medie tra i punti consecutivi lungo l'asse z")
diffs = [mean(abs.(z_values[i+1, :, :] .- z_values[i, :, :])) for i in 1:(length(z_values)-1)];
```

```julia
# trova gli indici dei due massimi
max_indices = partialsortperm(diffs, 1:2; rev=true)  # restituisce i due indici con i valori massimi

# calcola i nuovi punti come media dei due intervalli
new_points = [(z_values[i] + z_values[i+1]) / 2 for i in max_indices]

# aggiungi e ordina
z_values = sort!(vcat(z_values, new_points));
```
